# Fundamental Architecture

A 2-layer feedforward neural network has three layers: an input, a hidden layer, and an output layer.

The input's number of nodes is dependent on its number of features

- Input nodes ( = features): ~ 20 from estimation
- Hidden nodes: 12 (for now)
- Output nodes: 1 due to value being predicted

From [this post](https://stats.stackexchange.com/questions/181/how-to-choose-the-number-of-hidden-layers-and-nodes-in-a-feedforward-neural-netw) there many rules of thumb for estimating the right number of hidden nodes. From *Introduction to Neural Networks for Java* by Jeff Hayes we have

1. The number of hidden neurons should be between the size of the input layer and the size of the output layer.
2. The number of hidden neurons should be 2/3 the size of the input layer, plus the size of the output layer.
3. The number of hidden neurons should be less than twice the size of the input layer.

I'm going to select 12 as the number of hidden nodes to start, arbitrarily.

During training we should track nodes that have weights close to zero, because it is possible they can be pruned to improve performance.

For a fully connected hidden layer, there are $h$ nodes in the hidden layer and $i$ input features, with $s$ samples. The resultant weight matrix will be $h \times i$ in size. We have one bias per neuron, but add it to each sample so that it is $s \times h$ in size.

We have an input matrix of size $s \times i$ (columns are features, rows are samples). We multiply this with the weight matrix (transposed)

$$
\underbrace{X}_{s \times i}\underbrace{W_1^\top}_{h \times i = i \times h} +\underbrace{B_1}_{s \times h}=\underbrace{y_1}_{s \times h}
$$

We desire our output to have one feature, so it will be matrix of size $s \times o$.

$$
\underbrace{y_1}_{s \times h}\underbrace{W_2^\top}_{o \times h = h \times o} + \underbrace{B_2}_{s \times o} = \underbrace{y_2}_{s \times o}
$$

```python
def initalize_random(target_size, source_size):
    ini_mat = np.random.normal(loc=0.0, scale=1.0, size=(target_size, source_size)) # all weight matrices: # nodes of targeted layer x # nodes of source layer
    return ini_mat

def ReLU(x):
    relu = torch.maximum(torch.tensor(0.0), x) # removes negative values
    return relu

def forward(X, W_1, B_1, W_2, B_2):
    # input -> hidden
    hidden = torch.matmul(X, torch.transpose(torch.tensor(W_1))) + B_1 
    act_hidden = ReLU(hidden)
    # hidden -> output
    output = torch.matmul(act_hidden, torch.transpose(torch.tensor(W_2))) + B_2
    return output
```
